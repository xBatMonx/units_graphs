```python
for i in range(len(units)):
    if units['Unit Impedance'][i] == 'Out of Range':
        units['Unit Impedance'][i] = 10
    if units['Unit Voltage'][i] == 'Out of Range':
        units['Unit Voltage'][i] = 20
```

```python
num_units = 40
num_strings = 4

for k in range(1, num_strings+1):
    for j in range(1, num_units+1):
        print('unit', j)
        print('string', k)
        plot_imp_volt(units, j, k)
```

```python
def plot_imp_volt(units, unit, string):
    fig, ax1 = plt.subplots()
    ax1.plot([units['Unit Impedance'][i] for i in range(len(units)) if units['Unit'][i] == unit and units['String'][i] == string], 'b-')
    ax1.set_xlabel('time (weeks)')
    # Make the y-axis label, ticks and tick labels match the line color.
    ax1.set_ylabel('impedance', color='b')
    ax1.tick_params('y', colors='b')
    ax2 = ax1.twinx()
    ax2.plot([units['Unit Voltage'][i] for i in range(len(units)) if units['Unit'][i] == unit and units['String'][i] == string], 'r.')
    ax2.set_ylabel('voltage', color='r')
    ax2.tick_params('y', colors='r')
    fig.tight_layout()
    plt.show()
```
